```python
import random

def midpoint(p,q):
    return ((p[0]+q[0])/2,(p[1]+q[1])/2)

def d(p,q):
    x = p[0] - q[0]
    y = p[1] - q[1]
    return sqrt(x*x+y*y)

def draw_distance(p,q):
    rand_color = "#" + "".join([random.choice('23456789abc') for _ in range(6)])
    return line2d([p,q],color=rand_color) + text(f"{float(d(p,q)):.2f}",midpoint(p,q),color=rand_color)

lefts = ["#600","#a50","#fa0"]
rights = ["#006","#07a","#09f"]
```

```python
A = [[-1.0000,  0.0000],                                                                  
 [ 1.0000,  0.0000],                                                                  
 [-0.5732,  1.5583]]
B = [[-1.1603, -0.7836],                                                                  
         [-2.5716,  0.2701],                                                                  
         [-2.8240, -0.0433],                                                                  
         [-1.1178, -1.8315]]
```

```python
P = sum([point2d(a,color=color,size=50) for a,color in zip(A,lefts)]) + sum([point2d(a,color=color,size=50) for a,color in zip(B,rights)])
```

```python
P + sum([draw_distance(a,b) for a in A for b in B])
```

```python
def inverse_perm(P):
    Q = [0]*len(P)
    for i,p in enumerate(P):
        Q[p] = i
    return Q
```
